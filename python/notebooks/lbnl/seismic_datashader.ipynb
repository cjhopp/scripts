```python
import pandas as pd
import holoviews as hv
import datashader as ds
from holoviews.operation.datashader import datashade
import matplotlib.cm as mpl_cm
import matplotlib.colors as mpl_colors

hv.extension('bokeh')

# Load + parse times (existing)
df = pd.read_csv('/home/chopp/index_all_nanometrics.csv', usecols=['station','channel','starttime', 'archive_type'], dtype=str)
df['dt'] = pd.to_datetime(df['starttime'], format='%Y%m%d.%H%M%S', errors='coerce')
df['sta_chan'] = df['station'] + '_' + df['channel']
df = df.dropna(subset=['dt'])

# Limit to top N station+channel combinations for clarity (by count)
N = 200
top_sta_chan = df['sta_chan'].value_counts().index[:N].tolist()
df = df[df['sta_chan'].isin(top_sta_chan)].copy()

# helper to split and order by first char, station, channel
def split_sta_chan(s):
    parts = s.split('_', 1)
    sta = parts[0]
    chan = parts[1] if len(parts) > 1 else ''
    return sta, chan

# Sort by the uppercase first character (case-insensitive), then station, then channel
levels = sorted(
    top_sta_chan,
    key=lambda sc: (split_sta_chan(sc)[0][0].upper(), split_sta_chan(sc)[0].upper(), split_sta_chan(sc)[1])
)

df['sta_chan'] = pd.Categorical(df['sta_chan'], categories=levels, ordered=True)
df = df[df['sta_chan'].notna()].copy()
y_map = {lab: i for i, lab in enumerate(levels)}
df['y'] = df['sta_chan'].astype(str).map(y_map)

# derive first character (uppercase) for categorical colouring and grouping
df['first_char'] = df['station'].astype(str).str[0].str.upper()

# --- NEW: derive end time (dt_end) so we can draw bars ---
# Prefer explicit endtime/duration columns if present, otherwise use archive_type mapping
if 'endtime' in df.columns:
    df['dt_end'] = pd.to_datetime(df['endtime'], format='%Y%m%d.%H%M%S', errors='coerce')
elif 'dt_end' in df.columns:
    df['dt_end'] = pd.to_datetime(df['dt_end'], errors='coerce')
elif 'duration' in df.columns:
    df['dt_end'] = df['dt'] + pd.to_timedelta(df['duration'].astype(float), unit='s')
else:
    # Use archive_type if available: continuous -> 1800s, triggered -> 14s
    if 'archive_type' in df.columns:
        atype = df['archive_type'].astype(str).str.lower().fillna('')
        df['length_sec'] = atype.map({'continuous': 1800.0, 'triggered': 14.0}).fillna(1.0)
    else:
        # fallback single-second bars
        df['length_sec'] = 1.0
    df['dt_end'] = df['dt'] + pd.to_timedelta(df['length_sec'], unit='s')

# drop rows without valid start/end times
df = df.dropna(subset=['dt', 'dt_end']).copy()

# create vertical extents so each rectangle has thickness
bar_halfheight = 0.4   # tweak to make bars thicker/thinner (0.5 fills whole row)
df['y0'] = df['y'] - bar_halfheight
df['y1'] = df['y'] + bar_halfheight
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.colors as mpl_colors
import matplotlib.cm as mpl_cm
import matplotlib.dates as mdates

# Existing netcode_dict and color key logic
netcode_dict = {'B': 'Brady', 'D': 'Desert Peak', 'R': 'Raft River', 'F': 'Fallon', 'N': 'New York Canyon',
                '2': 'Patua', '4': 'Patua', '5': 'Patua'}

# Load the new file
new_data_file = "/home/chopp/station_dates.txt"  # Replace with the actual path
patua_data_file = "/home/chopp/Patua_dates.txt"
new_df = pd.read_csv(new_data_file, header=None, names=["station", "dt"])
new_df2 = pd.read_csv(patua_data_file, header=None, names=["station", "dt"], dtype={"station": str})
new_df = pd.concat([new_df, new_df2], ignore_index=True)
# Convert station names to uppercase and parse dates
new_df['station'] = new_df['station'].str.upper()
new_df['dt'] = pd.to_datetime(new_df['dt'])
# Assume each day corresponds to a full day of data
new_df['dt_end'] = new_df['dt'] + pd.Timedelta(days=1)

# Add a placeholder column for `first_char` (categorical coloring)
new_df['first_char'] = new_df['station'].str[0]
print(new_df)
```

```python

# Combine the new data with the existing DataFrame
df = pd.concat([df, new_df], ignore_index=True)

# color key by first_char
unique_chars = df['first_char'].dropna().unique()
cmap = mpl_cm.get_cmap('tab20', max(1, len(netcode_dict.keys())))
color_key = {net: mpl_colors.to_hex(cmap(i)) for i, (ch, net) in enumerate(netcode_dict.items())}

# Normalize station names in the combined DataFrame to ensure case-insensitive matching
df['station'] = df['station'].str.upper()

# Drop stations starting with "ACQ" (if needed)
df = df[~df['station'].str.startswith('ACQ', na=False)].copy()

# Recalculate levels for stations (ensure all stations are included)
stations = sorted(df['station'].unique())
df['station'] = pd.Categorical(df['station'], categories=stations, ordered=True)
y_map = {lab: i for i, lab in enumerate(stations)}
df['y'] = df['station'].astype(str).map(y_map)

# Aggregate data into weekly bins (but keep absolute datetimes)
df['week_start'] = df['dt'].dt.to_period('W').apply(lambda r: r.start_time)  # Start of the week
weekly_df = df.groupby(['week_start', 'station', 'first_char']).agg(
    y=('y', 'first'),  # Use the first y value for the station
    count=('dt', 'count')  # Count the number of patches in the week
).reset_index()

weekly_df.to_json("seismic_availability.json", orient='records', date_format='iso')

# df = pd.read_json("seismic_availability.json", convert_dates=['dt', 'dt_end'])

# Create vertical extents for the aggregated bars
bar_halfheight = 0.4
weekly_df['y0'] = weekly_df['y'] - bar_halfheight
weekly_df['y1'] = weekly_df['y'] + bar_halfheight
weekly_df['x0'] = weekly_df['week_start']
weekly_df['x1'] = weekly_df['week_start'] + pd.Timedelta(weeks=1)  # Each bin spans one week

# Plot the aggregated data
fig, ax = plt.subplots(figsize=(16, max(6, len(stations) * 0.2)))

for _, row in weekly_df.iterrows():
    color = color_key.get(netcode_dict[row['first_char']], '#000000')  # Default to black
    rect = patches.Rectangle(
        (row['x0'], row['y0']),  # Bottom-left corner
        row['x1'] - row['x0'],  # Width (1 week)
        row['y1'] - row['y0'],  # Height
        color=color,
        alpha=0.8,
        edgecolor='black',
        linewidth=0.5
    )
    ax.add_patch(rect)

# Set axis limits and labels
ax.set_xlim(weekly_df['x0'].min(), weekly_df['x1'].max())
ax.set_ylim(-0.5, len(stations) - 0.5)
ax.set_xlabel("Time", fontsize=16)
ax.set_ylabel("Station", fontsize=16)
ax.set_yticks(range(len(stations)))
ax.set_yticklabels(stations, fontsize=12)

# Format x-axis to display datetime labels
ax.xaxis.set_major_locator(mdates.YearLocator())  # Major ticks every year
ax.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=3))  # Minor ticks every 3 months
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # Format as "YYYY-MM"

# Rotate x-axis labels for better readability
plt.xticks(rotation=20, fontsize=16)

# Add a legend for the color categories
handles = [
    patches.Patch(color=color_key[netcode_dict[ch]], label=netcode_dict.get(ch, ch))
    for ch in unique_chars if ch not in ['A', '4', '5']
]
ax.legend(handles=handles, title="Station Group", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=14)

# Add a grid for better readability
ax.grid(True, linestyle='--', alpha=0.5)

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure as a high-resolution image
plt.savefig("seismic_availability_with_new_stations.png", dpi=300)

# Show the plot
plt.show()
```
