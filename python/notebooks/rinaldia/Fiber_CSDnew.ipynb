```python
import numpy as np
import os
import datetime
from matplotlib.ticker import ScalarFormatter
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib as mpl
```

```python
def load_data(filename):
    f=open(filename,'rb')
    data_time=np.genfromtxt(f,delimiter="\t",dtype=np.str,skip_header=1,max_rows=1)
    length=np.genfromtxt(f,delimiter="\t",dtype=np.float,skip_header=1,max_rows=1)
    data=np.genfromtxt(f,delimiter=" ",dtype=np.float,skip_header=1)
    time=np.zeros(len(data_time),dtype='datetime64[s]')
    for i in range(0,len(data_time)):
        time[i-1]=datetime.datetime.strptime(data_time[i], "%d/%m/%Y %H:%M:%S")
    
    return length,time,data
    
```

```python
def load_data2(filename):
    f=open(filename,'rb')
    data_time=np.genfromtxt(f,delimiter="\t",dtype=np.str,skip_header=1,max_rows=1)
    length=np.genfromtxt(f,delimiter="\t",dtype=np.float,skip_header=1,max_rows=1)
    data=np.genfromtxt(f,delimiter="\t",dtype=np.float,skip_header=1)
    time=np.zeros(len(data_time)-1,dtype='datetime64[s]')
    for i in range(0,len(data_time)-1):
        time[i-1]=datetime.datetime.strptime(data_time[i], "%d/%m/%Y %H:%M:%S")
    
    return length,time,data
```

```python
BCSD3_len1,BCSD3_time1,BCSD3_data1=load_data('../data_formatted_new/BCSD3_down.out')
BCSD3_len2,BCSD3_time2,BCSD3_data2=load_data('../data_formatted_new/BCSD3_up.out')

BCSD4_len1,BCSD4_time1,BCSD4_data1=load_data('../data_formatted_new/BCSD4_down.out')
BCSD4_len2,BCSD4_time2,BCSD4_data2=load_data('../data_formatted_new/BCSD4_up.out')

BCSD5_len1,BCSD5_time1,BCSD5_data1=load_data('../data_formatted_new/BCSD5_down.out')
BCSD5_len2,BCSD5_time2,BCSD5_data2=load_data('../data_formatted_new/BCSD5_up.out')

BCSD6_len1,BCSD6_time1,BCSD6_data1=load_data('../data_formatted_new/BCSD6_down.out')
BCSD6_len2,BCSD6_time2,BCSD6_data2=load_data('../data_formatted_new/BCSD6_up.out')
```

```python
extBCSD5_len1,extBCSD5_time1,extBCSD5_data1=load_data2('../data_formatted_new/extBCSD5.out')
extBCSD5_len2,extBCSD5_time2,extBCSD5_data2=load_data2('../data_formatted_new/extBCSD5_short.out')
```

## Borehole CSD-3

```python
font = {'size'   : 22}

mpl.rc('font', **font)

fig, axes = plt.subplots(figsize=(20, 7), nrows=1, ncols=2)
formatter = ScalarFormatter(useMathText=True)
formatter.set_scientific(True)

cset1 = axes[0].contourf(BCSD3_time1, BCSD3_len1, \
                           BCSD3_data1,levels=np.linspace(-500,500,21),cmap="RdBu_r")
fig.colorbar(cset1, ax=axes[0])

axes[0].xaxis.set_major_formatter(
        mpl.dates.DateFormatter('%d-%m-%y')
        )
axes[0].yaxis.set_major_formatter(formatter)
axes[0].xaxis.set_major_locator(ticker.FixedLocator(\
                                    [737202., 737204., 737206., 737208., 737210., 737212., 737214.]))
ylim = axes[0].get_ylim()
axes[0].set_ylim(ylim[::-1])
axes[0].set_xlabel('Time (days)')
axes[0].set_ylabel('Fiber length (m)')
axes[0].set_title('BCSD3 (DOWN)')

cset2 = axes[1].contourf(BCSD3_time2, BCSD3_len2, \
                           BCSD3_data2,levels=np.linspace(-500,500,21),cmap="RdBu_r")
fig.colorbar(cset2, ax=axes[1])


axes[1].xaxis.set_major_formatter(
        mpl.dates.DateFormatter('%d-%m-%y')
        )
axes[1].yaxis.set_major_formatter(formatter)
axes[1].xaxis.set_major_locator(ticker.FixedLocator(\
                                    [737202., 737204., 737206., 737208., 737210., 737212., 737214.]))
ylim = axes[1].get_ylim()
axes[1].set_ylim(ylim[::-1])
axes[1].set_xlabel('Time (days)')
axes[1].set_ylabel('Fiber length (m)')
axes[1].set_title('BCSD3 (UP)')

plt.gcf().autofmt_xdate()
fig.subplots_adjust(hspace = 0.8,left=0.14,right=0.90)


plt.show()
```
