```python
import gmt
import numpy as np

from glob import glob
from obspy import read_events, UTCDateTime
from gmt.clib import LibGMT
from gmt.base_plotting import BasePlotting
from gmt.utils import build_arg_string, dummy_context, data_kind
```

```python
cat_Nga_temps_DD = read_events('/media/chet/hdd/seismic/NZ/hypoDD/all_temps_Nga_7_spicks_manual/det_cat_mcc0.5_shift0.2_ALL_TEMPS_NGA_HypoDD_cc0.7_Spicks_manual_just_hypoDD.xml')
```

```python
# Set x-section params
c_lon = 176.190
c_lat = -38.546
e_lon = 176.209
e_lat = -38.575
s_lon = e_lon - (2 * (e_lon - c_lon))
s_lat = e_lat - (2 * (e_lat - c_lat))
print((s_lon, s_lat))
max_depth = 4.5
```

```python
# Set up location arrays
lats = np.asarray([ev.origins[-1].latitude for ev in cat_Nga_temps_DD])
lons = np.asarray([ev.origins[-1].longitude for ev in cat_Nga_temps_DD])
print(lons)
# Make array of elapsed seconds since start of catalog and normalized magnitudes
dates = [ev.origins[-1].time.datetime for ev in cat_Nga_temps_DD]
dates.sort()
secs = [(d - dates[0]).total_seconds() for d in dates]
secs /= np.max(secs)
# Magnitude scaling as well
mags = [ev.magnitudes[-1].mag for ev in cat_Nga_temps_DD]
mags /= np.max(mags)
```

```python
# Make map view #
# Start new figure
fig = gmt.Figure()
# Set preferences
with LibGMT() as lib:
    lib.call_module('gmtset', 'FORMAT_GEO_MAP ddd.xx')
    lib.call_module('gmtset', 'MAP_FRAME_TYPE plain')
# Plot resistivity
region = [176.15, 176.23, -38.58, -38.51]
fig.plot(region=region, projection='M13', Y='10',
         data='/home/chet/gmt/data/NZ/resistivity/RT_Boundary_Risk_2000.gmt', color='lightgray')
fig.plot(data='/home/chet/gmt/data/NZ/resistivity/NM_Boundary_Bosely2010.gmt', color='lightgray')
# Water
fig.plot(data='/home/chet/gmt/data/NZ/water/taupo_lakes.gmt', color='cornflowerblue')
fig.plot(data='/home/chet/gmt/data/NZ/water/taupo_river_polygons.gmt', color='cornflowerblue')
# Faults
fig.plot(data='/home/chet/gmt/data/NZ/faults/NZ_active_faults.gmt', W='0.85')
# Hypocenters
fig.plot(color=secs, cmap='cool', x=lons, y=lats, sizes=mags/4, style='cc')
# Plot well tracks map view
fig.plot(data='/home/chet/gmt/data/NZ/wells/NM_tracks_injection.gmt', W='1.5,blue', S='qn1:+f8p,blue+Lh')
fig.plot(data='/home/chet/gmt/data/NZ/wells/NM_tracks_production.gmt', W='1.5,red', S='qn1:+f8p,red+Lh')
fig.plot(data='/home/chet/gmt/data/NZ/wells/NM06_track.gmt', W='1.5,blue', S='qn1:+f8p,blue+Lh+n-.2/0.2')
fig.plot(data='/home/chet/gmt/data/NZ/wells/NM08_track.gmt', W='1.5,blue', S='qn1:+f8p,blue+Lh+n0/-0.15')
fig.plot(data='/home/chet/gmt/data/NZ/wells/NM10_track.gmt', W='1.5,blue', S='qn1:+f8p,blue+Lh+n0/0')
fig.plot(data='/home/chet/gmt/data/NZ/wells/NM09_track.gmt', W='1.5,blue', S='qn1:+f8p,blue+Lh+n0/0')
fig.plot(data='/home/chet/gmt/data/NZ/wells/NM12_track.gmt', W='1.5,red', S='qn1:+f8p,red+Lh')
# Wellhead points
fig.plot(data='/home/chet/gmt/data/NZ/wells/NM_injection_pts.gmt', color='blue', style='c0.2c')
fig.plot(data='/home/chet/gmt/data/NZ/wells/NM_production_pts.gmt', color='red', style='c0.2c')
# Stations
fig.plot(data='/home/chet/gmt/data/NZ/stations/mrp_inventory.gmt', style='t0.3c', color='black', W='0.07')
fig.basemap(region=region, projection='M13', L='x11.5/1/-38/1',frame=['0.02','wsEN'])
```

```python
# Trying some pstext action with clib calls
pts5_7 = '/home/chet/gmt/data/NZ/wells/NM05_NM07_pts.gmt'
args5_7 = {'D': '0.25/0.15', 'F': '+jCB+f8p,red'}
pt4 = '/home/chet/gmt/data/NZ/wells/NM04_pt.gmt'
args4 = {'D': '0.35/-0.35'}
kwargs = args5_7
with LibGMT() as lib:
    file_context = dummy_context(pts5_7)
    with file_context as fname:
        arg_str = ' '.join([fname, build_arg_string(kwargs)])
        lib.call_module('pstext', arg_str)
# Try plotting the scale
scale_args = {'C': 'cool', 'D': '10.25/13.75/3.5/0.5h', 'B': 'px1+l"Date of occurrence"',
              'F': '+gwhite+p1.2k'}
with LibGMT() as lib:
    lib.call_module('psscale', ' '.join([build_arg_string(scale_args),
                                         '--FONT_ANNOT_PRIMARY=7p', '--FONT_LABEL=9p']))
# Cross-section line and labels
fig.plot(x=np.asarray([s_lon, e_lon]), y=np.asarray([s_lat, e_lat]), W='1.0,black,-')
labels_file = '/home/chet/gmt/data/NZ/misc/section_labels.txt'
with LibGMT() as lib:
    label_args = {'F': '+f16p,black,-'}
    file_context = dummy_context(labels_file)
    with file_context as fname:
        arg_str = ' '.join([fname, build_arg_string(label_args)])
        lib.call_module('pstext', arg_str)
```

```python
# Cross section
scale='X13/-6'
# Set preferences
with LibGMT() as lib:
    lib.call_module('gmtset', 'FONT_ANNOT_PRIMARY 10p')
    lib.call_module('gmtset', 'FONT_LABEL 12p')
proj_evs = []
with open('/home/chet/gmt/data/NZ/locations/hypoDD/outfile_temps_Nga.tmp', 'r') as f:
    for line in f:
        proj_evs.append([float(line.split('\t')[0]), float(line.split('\t')[1])])
x, y = zip(*proj_evs)
fig.plot(x=np.array(x), y=np.array(y), sizes=mags/4, color=secs, J=scale, Y='-7.25', cmap='cool', style='cc',
         R='-3.635/3.635/-0.5/{}'.format(max_depth), B=['px1.0+l"Distance from center of x-section (km)"',
                                                        'py1.0+l"Depth (km)"', 'WeSn'])
proj_wells = glob('/home/chet/gmt/data/NZ/wells/*project.txt')
for well in proj_wells:
    well_nm = well.split('/')[-1].split('_')[0]
    if well_nm in ['NM10', 'NM06', 'NM08', 'NM09']:
        fig.plot(data=well, W='0.9p,blue')
    else:
        fig.plot(data=well, W='0.9p,red')
proj_fzs = glob('/home/chet/gmt/data/NZ/wells/feedzones/*project.csv')
for well in proj_fzs:
    fig.plot(data=well, W='6.5p,blue')
```
